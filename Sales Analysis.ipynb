```python
# import python libraries


# NumPy is used for numerical operations and array handling
import numpy as np 
# Pandas is used for data manipulation and analysis
import pandas as pd 
# Matplotlib is a plotting library for creating visualizations
import matplotlib.pyplot as plt
# The following line is a Jupyter Notebook magic command to display plots within the notebook
%matplotlib inline
# Seaborn is a statistical data visualization library based on Matplotlib
import seaborn as sns
```

```python
# Used the pd.read_csv() function to read the CSV file 'Sales Data.csv'
# Specified the encoding parameter as 'unicode_escape' to handle any Unicode encoding issues

df = pd.read_csv('Sales Data.csv', encoding= 'unicode_escape')
```

# Data Cleaning

```python
# Drop the columns named 'Status' and 'unnamed1'
# axis=1 specifies that we are dropping columns, not rows
# inplace=True means the changes are applied directly to the original DataFrame 'df'

df.drop(['Status','unnamed1'], axis=1, inplace=True)
```

```python
# The 'isnull()' function is used to check for missing or null values in each cell of the DataFrame.

df.isnull().sum()
```

```python
# To remove rows with missing values.

df.dropna(inplace=True)
```

```python
# df.isnull() returns a DataFrame of the same shape as df, where each element is True if the corresponding element in df is Null
df.isnull().sum()
```

```python
#Convert the 'Amount' column in the DataFrame to integer type

df['Amount'] = df['Amount'].astype('int')
```

# Exploratory Data Analysis

### Gender

```python
df.columns
```

```python
# plotting a bar chart for Gender and it's count

# Create a countplot using Seaborn
ax = sns.countplot(x = 'Gender',data = df)
# Iterate over the bars in the countplot
for bars in ax.containers:
    # Add labels to the bars based on the height of each bar
    ax.bar_label(bars)
```

```python
# Group the DataFrame by the 'Gender' column
# 'as_index=False' ensures that 'Gender' is not set as the index in the resulting DataFrame
# The grouped data will be aggregated using a sum operation on the 'Amount' column
# 'ascending=False' means sorting in descending order

df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
```

```python
# plotting a bar chart for gender vs total amount

sales_gen = df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.barplot(x = 'Gender',y= 'Amount' ,data = sales_gen)

ax = sns.barplot(x='Gender', y='Amount', data=sales_gen)
for index, value in enumerate(sales_gen['Amount']):
    ax.text(index, value, str(value), ha='center', va='bottom')
```

*The data depicted in the graph indicates a higher proportion of female buyers, with females exhibiting greater purchasing power compared to their male counterparts.*

### AGE

```python
df.columns
```

```python
ax = sns.countplot(data = df, x = 'Age Group')
```

```python
# 'hue' parameter specifies the variable to differentiate by color, which is 'Gender' in this case

ax = sns.countplot(data = df, x = 'Age Group', hue = 'Gender')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# Total Amount vs Age Group
sales_age = df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.barplot(x = 'Age Group',y= 'Amount' ,data = sales_age)
```

*The graphs indicate that a predominant portion of purchasers falls within the 26 to 35 years age bracket, with a majority being female.*

### State

```python
df.columns
```

```python
# total number of orders from top 15 states

sales_state = df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(15)
# 'figure.figsize' specifies the width and height of the figure in inches
sns.set(rc={'figure.figsize':(25,10)})
sns.barplot(data = sales_state, x = 'State',y= 'Orders')
```

```python
# total amount/sales from top 15 states

sales_state = df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(15)
# 'figure.figsize' specifies the width and height of the figure in inches
sns.set(rc={'figure.figsize':(25,10)})
sns.barplot(data = sales_state, x = 'State',y= 'Amount')
```

*The graphs indicate that a significant portion of the orders and total sales/amount is concentrated in Uttar Pradesh, Maharashtra, and Karnataka, respectively.*

### Marital Status

```python
df.columns
```

```python
ax = sns.countplot(data = df, x = 'Marital_Status')
# Set the size of the figure using rc (runtime configuration) parameters
sns.set(rc={'figure.figsize':(5,8)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state = df.groupby(['Marital_Status', 'Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.set(rc={'figure.figsize':(5,5)})
sns.barplot(data = sales_state, x = 'Marital_Status',y= 'Amount', hue='Gender')
```

*Based on the aforementioned charts, it is evident that a predominant portion of the purchasers comprises married individuals, predominantly women, who exhibit a considerable degree of economic influence.*

### Occupation

```python
sns.set(rc={'figure.figsize':(20,5)})
ax = sns.countplot(data = df, x = 'Occupation')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state = df.groupby(['Occupation'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_state, x = 'Occupation',y= 'Amount')
```

*Based on the charts presented earlier, it is evident that a significant portion of purchasers are employed in the information technology, healthcare, and aviation industries.*

### Product Category

```python
sns.set(rc={'figure.figsize':(25,10)})
ax = sns.countplot(data = df, x = 'Product_Category')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state = df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_state, x = 'Product_Category',y= 'Amount')
```

*Upon examining the aforementioned charts, it becomes apparent that a significant portion of the merchandise sold belongs to the Food, Apparel, and Electronics sectors.*

```python
# Most selling products

sales_state = df.groupby(['Product_ID'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_state, x = 'Product_ID',y= 'Orders')
```

### Conclusion:

*Women between the ages of 26 and 35, who are married and employed in the Information Technology (IT), Healthcare, and Aviation sectors in the states of Uttar Pradesh (UP), Maharashtra, and Karnataka, exhibit a heightened propensity to make purchases across the spectrum of Food, Clothing, and Electronics categories.*
